```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


import yfinance as yf
import pyfolio as pf
```

Let's analyze TCS stock performance. As of Apr-June 2026, this stock (and all TCS IT Services stocks) have been doing badly in the "AI-for-everything" wave.

**Step 1:** Download the stock data [TCS stock symbol is `TCS.NS` on Yahoo Finance!]. We'll download daily stock prices - all available data. It will return the data as a Pandas `DataFrame`, with Open/High/Low/Close prices + Volume, along with other fields.

```python
# get me all the available stock data (history("max"))
data = yf.Ticker("TCS.NS").history("max")
data.tail(5)
```

**Step 2:** Let's calculate the **daily stock returns**, which is a measure of _how much money the stock **gained or lost** from one day to the next_, expressed as a percentage. Pandas provides a `pct_change()` function, which is applied to the Closing price of the stock. The `pct_change()` function returns a float, not %, so I multiply by 100 - a -ve sign means down and +ve sign means up.

```python
returns = data.Close.pct_change() * 100
returns.tail(10)  # view latest returns
```

```python
# plot the last 30 day returns to see how it trends
# Slice and plot directly
def plot_returns(returns: pd.Series, days: int = 30) -> None:
    returns.tail(days).plot(
        figsize=(12, 6),
        grid=True,
        title=f"Stock Daily returns for last {days} days (%)",
        ylabel="Return (%)",
    )


plot_returns(returns, 50)
```
